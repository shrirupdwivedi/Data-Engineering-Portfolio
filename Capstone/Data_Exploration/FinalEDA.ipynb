# Businss Dataset

```python
import pandas as pd
import seaborn as sns
import json
import pprint
import ast
```

The JSON file is converted to Pandas DataFrame in order for the sake of understanding and ETL process. The data is loaded in chunks to decrease memory load on the system.

```python
filename = '/Volumes/Sadhguru/yelp-dataset/yelp_academic_dataset_business.json'
for df in pd.read_json(filename,chunksize=1000,lines=True):
    df_b = df
display(df_b.head())

    
```

```python
df_b['categories'][150006]
```

```python
df_b.isnull().sum()
```

The above result shows that attributes and hours columns in the dataset have at lease one null value. The null values in attribute column will be deleted as it would create problems when quering the data later in the analysis stage.

```python
df_b1=df_b.dropna(subset=['attributes'])
```

```python
display(df_b1)
```

The categories column seems to have many items. Expanding on the column would give us better insight.

```python
attr = df_b1['attributes'][150000]
pprint.pprint(attr)
```

It can be seen that the attribute column gives many useful insights for a business.


As we are just interested in Restaurants, other categories of businesses should be filtered out in the categories column

```python
df_b_restaurants = df_b1[df_b1["categories"].str.contains("Restaurants") == True]
```

```python
display(df_b_restaurants.head())
```
