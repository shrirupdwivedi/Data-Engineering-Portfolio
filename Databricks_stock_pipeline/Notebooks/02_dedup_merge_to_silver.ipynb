```python
from pyspark.sql.functions import row_number, col, current_timestamp
from pyspark.sql.window import Window
dbutils.widgets.text("trade_date", "")
trade_date = dbutils.widgets.get("trade_date")
```

```python
df_raw = spark.read.format("delta").load("/Volumes/capstone/default/raw_market_data")
```

```python
window_spec = Window.partitionBy("trade_dt", "rec_type", "symbol", "event_tm", "event_seq_nb", "exchange").orderBy(col("arrival_tm").desc())
df_dedup = df_raw.withColumn("rank", row_number().over(window_spec)).filter(col("rank") == 1).drop("rank").withColumn("merge_ts", current_timestamp())
display(df_dedup)
```

```python
df_trade = df_dedup.filter("rec_type = 'T'")
df_quote = df_dedup.filter("rec_type = 'Q'")
df_trade.write.format("delta") \
    .mode("overwrite") \
    .save("/Volumes/capstone/default/trade_data")
df_quote.write.format("delta") \
    .mode("overwrite") \
    .save("/Volumes/capstone/default/quote_data")

df_trade.createOrReplaceTempView("source_trade")
df_quote.createOrReplaceTempView("source_quote")
```

```python
%sql
MERGE INTO delta.`/Volumes/capstone/default/trade_data` as target
USING source_trade as source
ON target.trade_dt = source.trade_dt AND
    target.rec_type = source.rec_type AND
    target.symbol = source.symbol AND
    target.exchange = source.exchange AND
    target.event_tm = source.event_tm AND
    target.event_seq_nb = source.event_seq_nb
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *
```

```python
%sql
MERGE INTO delta.`/Volumes/capstone/default/quote_data` AS target
USING source_quote AS source
ON  target.trade_dt = source.trade_dt AND
    target.rec_type = source.rec_type AND
    target.symbol = source.symbol AND
    target.exchange = source.exchange AND
    target.event_tm = source.event_tm AND
    target.event_seq_nb = source.event_seq_nb
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *
```
